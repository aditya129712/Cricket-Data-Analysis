```python
import pandas as pd
import numpy as np
```

```python
dataset = pd.read_csv('match_export.csv')
dataset.head()
```

```python
dataset['RESULT'].unique()
```

```python
dataset['RESULT'] = dataset['RESULT'].map({'won': 1,'lost':0,'n/r':2,'aban':3,'tied':4,'canc':5,'0':6})
```

```python
dataset
```

```python
dataset['TOSS'].unique()
```

```python
dataset['TOSS'] = dataset['TOSS'].map({'lost':0, 'won':1, '0':2})
dataset
```

```python
dataset['BAT'].unique()
```

```python
dataset['BAT'] =dataset['BAT'].map({'1st':1, '2nd':2, '0':3})
dataset
```

```python
dataset['MARGIN'] = dataset['MARGIN'].str.split(' ').str[0]
dataset
```

```python
dataset['MARGIN'] = dataset['MARGIN'].astype(int)
dataset
```

```python
dataset.info()
```

```python
final_df_match = pd.DataFrame(dataset)
final_df_match
```
